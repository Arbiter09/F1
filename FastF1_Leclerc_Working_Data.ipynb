```python
import fastf1
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
from IPython.display import display

# Define cache directory (adjust to your environment)
cache_dir = 'On My iPad/Carnets/f1_cache'

# Ensure the cache directory exists
if not os.path.exists(cache_dir):
    os.makedirs(cache_dir)

# Enable FastF1 cache
fastf1.Cache.enable_cache(cache_dir)

# Load a specific race session (2021 Abu Dhabi GP)
race_year = 2021
race_name = 'Abu Dhabi'
session_type = 'R'  # R for Race, Q for Qualifying, FP for Free Practice

# Load the session data
session = fastf1.get_session(race_year, race_name, session_type)
session.load()

# Filter data for Charles Leclerc
laps = session.laps.pick_driver('16')  # Charles Leclerc's driver number is '16'

# Ensure that there are valid laps for Charles Leclerc
if laps.empty:
    raise ValueError("No lap data found for Charles Leclerc.")

# Extract telemetry data for the fastest lap of Charles Leclerc
lap = laps.pick_fastest()  # Selecting the fastest lap for detailed telemetry
leclerc_telemetry = lap.get_telemetry()

# Clean telemetry data to handle missing values
leclerc_telemetry = leclerc_telemetry.fillna(method='ffill').fillna(method='bfill')

# Get the session start time
session_start_time = session.date  # This is the session start time as a datetime object

# Calculate the actual LapStartTime by adding the timedelta to the session start time
laps['LapStartTime'] = session_start_time + laps['LapStartTime']

# Load weather data and adjust time using session start time
weather_data = session.weather_data[['Time', 'AirTemp', 'TrackTemp', 'Rainfall']]
weather_data['Time'] = session_start_time + weather_data['Time']  # Adjust weather time

# Merge the weather data with lap data based on the nearest time
laps_with_weather = pd.merge_asof(laps, weather_data, left_on='LapStartTime', right_on='Time', direction='backward')

# 3. Tire temperatures (Check availability)
tire_temp_data = pd.DataFrame()
if all(col in laps.columns for col in ['TyreTemperatureFrontLeft', 'TyreTemperatureFrontRight', 'TyreTemperatureRearLeft', 'TyreTemperatureRearRight']):
    tire_temp_data = laps[['LapNumber', 'TyreTemperatureFrontLeft', 'TyreTemperatureFrontRight',
                           'TyreTemperatureRearLeft', 'TyreTemperatureRearRight']]

# Ensure 'LapNumber' exists in tire_temp_data for merging
if not tire_temp_data.empty and 'LapNumber' not in tire_temp_data.columns:
    tire_temp_data['LapNumber'] = laps['LapNumber']

# 4. Driver style: Extract more detailed telemetry data
required_columns = ['Distance', 'Speed', 'Brake', 'Throttle', 'nGear', 'SteerAngle']
available_telemetry_columns = leclerc_telemetry.columns
driver_style_data = pd.DataFrame()

if all(col in available_telemetry_columns for col in required_columns):
    driver_style_data = leclerc_telemetry[required_columns]
else:
    available_columns_to_extract = [col for col in required_columns if col in available_telemetry_columns]
    if available_columns_to_extract:
        driver_style_data = leclerc_telemetry[available_columns_to_extract]

# 5. Grip levels (inferred from tire temperature and compound)
if not tire_temp_data.empty and 'TrackTemp' in laps_with_weather.columns:
    tire_temp_data['AvgTireTemp'] = tire_temp_data[['TyreTemperatureFrontLeft', 'TyreTemperatureFrontRight',
                                                    'TyreTemperatureRearLeft', 'TyreTemperatureRearRight']].mean(axis=1)
    grip_levels = pd.DataFrame()
    grip_levels['LapNumber'] = tire_temp_data['LapNumber']
    grip_levels['GripLevel'] = tire_temp_data['AvgTireTemp'] - laps_with_weather['TrackTemp']  # Simplified grip metric
else:
    grip_levels = pd.DataFrame()

# Ensure 'LapNumber' exists in both DataFrames before merging
if 'LapNumber' in laps_with_weather.columns and 'LapNumber' in tire_temp_data.columns:
    # Create a comprehensive DataFrame with all the factors for Charles Leclerc
    all_factors = pd.merge(laps_with_weather, tire_temp_data, on='LapNumber', how='left')
else:
    all_factors = laps_with_weather  # Fallback to laps_with_weather if tire_temp_data is empty or lacks LapNumber

# If grip_levels exist, merge them too
if not grip_levels.empty:
    all_factors = pd.merge(all_factors, grip_levels, on='LapNumber', how='left')

# Display the final DataFrame with all factors
print("All Factors for Charles Leclerc:")
display(all_factors)

# ---- Save the DataFrame to CSV ---- #
csv_file_path = 'charles_leclerc_race_data.csv'  # Define the CSV file name and path
all_factors.to_csv(csv_file_path, index=False)   # Save DataFrame to CSV without row index
print(f"Data saved to {csv_file_path}")

# Plotting visualizations for each factor (only if data is available)

# 1. Tire Age over Laps
plt.figure(figsize=(10, 5))
plt.plot(all_factors['LapNumber'], all_factors['TyreLife'], marker='o', linestyle='-', color='b')
plt.xlabel('Lap Number')
plt.ylabel('Tire Age (Laps)')
plt.title('Tire Age Over Laps for Charles Leclerc')
plt.grid(True)
plt.show()

# 2. Track Temperature and Air Temperature over Laps
if 'TrackTemp' in all_factors.columns and 'AirTemp' in all_factors.columns:
    plt.figure(figsize=(10, 5))
    plt.plot(all_factors['LapNumber'], all_factors['TrackTemp'], label='Track Temperature', color='r')
    plt.plot(all_factors['LapNumber'], all_factors['AirTemp'], label='Air Temperature', color='g')
    plt.xlabel('Lap Number')
    plt.ylabel('Temperature (°C)')
    plt.title('Track and Air Temperature Over Laps for Charles Leclerc')
    plt.legend()
    plt.grid(True)
    plt.show()

# 3. Rainfall over Laps (Weather Conditions)
if 'Rainfall' in all_factors.columns:
    plt.figure(figsize=(10, 5))
    plt.plot(all_factors['LapNumber'], all_factors['Rainfall'], label='Rainfall', color='blue')
    plt.xlabel('Lap Number')
    plt.ylabel('Rainfall (mm)')
    plt.title('Rainfall Over Laps for Charles Leclerc')
    plt.legend()
    plt.grid(True)
    plt.show()

# 4. Driver Style: Brake and Throttle over Distance (if available)
if 'Brake' in driver_style_data.columns and 'Throttle' in driver_style_data.columns:
    plt.figure(figsize=(10, 5))
    plt.plot(driver_style_data['Distance'], driver_style_data['Brake'], label='Brake', color='red')
    plt.plot(driver_style_data['Distance'], driver_style_data['Throttle'], label='Throttle', color='blue')
    plt.xlabel('Distance (m)')
    plt.ylabel('Brake/Throttle (%)')
    plt.title('Brake and Throttle Usage Over Track Distance for Charles Leclerc')
    plt.legend()
    plt.grid(True)
    plt.show()

# 5. Track Position over Laps
plt.figure(figsize=(10, 5))
plt.plot(all_factors['LapNumber'], all_factors['Position'], marker='o', linestyle='-', color='magenta')
plt.xlabel('Lap Number')
plt.ylabel('Track Position')
plt.title('Track Position Over Laps for Charles Leclerc')
plt.gca().invert_yaxis()  # Lower values mean better position
plt.grid(True)
plt.show()

# 6. Gap to Leader over Laps (updated code)
if 'GapToLeader' in all_factors.columns:
    if all_factors['GapToLeader'].notna().any():
        plt.figure(figsize=(10, 5))
        plt.plot(all_factors['LapNumber'], all_factors['GapToLeader'], marker='o', linestyle='-', color='cyan')
        plt.xlabel('Lap Number')
        plt.ylabel('Gap to Leader (s)')
        plt.title('Gap to Leader Over Laps for Charles Leclerc')
        plt.grid(True)
        plt.show()
    else:
        print("No Gap to Leader data available.")
else:
    print("Gap to Leader column is not available in the data.")
```
